```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Reading the CSV file and creating a data frame
car_specs_df = pd.read_csv('Car_specifications.csv')

# Reading the Excel file and creating a data frame
car_prices_df = pd.read_csv('Car_Prices.csv')
```

```python
# Displaying the data frame car_specs_df
car_specs_df
```

```python
# Displaying the data frames car_prices_df
car_prices_df
```

```python
# Merging the two data frames on the basis of 'Model Name' column
merged_df = pd.merge(car_specs_df, car_prices_df, on='Model Name')
merged_df
```

```python
# Filling the 'NA' or missing values with 0
merged_df.fillna(0, inplace=True)
```

```python
# Result after filling 'NA' vaues with 0
merged_df
```

```python
# Iterating over the merged data frame and adding GST value in 'On road pricing' column
for index, row in merged_df.iterrows():
    on_road_price = row['On road pricing']
    gst_value = on_road_price * 0.18
    merged_df.at[index, 'On road pricing'] = on_road_price + gst_value
```

```python
# Result after updating 
merged_df
```
