```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Admission_Predict.csv")
```

```python
df
```

```python
df.columns
```

```python
df=df.drop('Serial No.',axis=1)
```

```python
df
```

```python
df['admission']=df['Chance of Admit ']>0.75
```

```python
df
```

```python
df=df.replace({True:1,False:0})
```

```python
df
```

```python
df1=df.drop('Chance of Admit ',axis=1)
```

```python
df1
```

```python
df1.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
y=df1['admission']
y
```

```python
df1=df1.drop('admission',axis=1)
df1
```

```python
train_x,test_x,train_y,test_y=train_test_split(df1,y,test_size=0.25)
```

```python
train_x
```

```python
train_y
```

```python
test_x
```

```python
test_y
```

```python
# criterion is default=”gini”
dt=DecisionTreeClassifier()
dt=dt.fit(train_x,train_y)
```

```python
plot_tree(dt)
plt.savefig('tree.pdf')
plt.show()
```

```python
p=dt.predict(train_x)
```

```python
confusion_matrix(train_y,p)
```

```python
accuracy_score(train_y,p)*100
```

```python
p1=dt.predict(test_x)
```

```python
confusion_matrix(test_y,p1)
```

```python
accuracy_score(test_y,p1)*100
```

```python
# now criterion is=“entropy”
dt1=DecisionTreeClassifier(criterion="entropy")
dt1=dt1.fit(train_x,train_y)
```

```python
plot_tree(dt1)
plt.savefig('tree_entropy.pdf')
plt.show()
```

```python
p2=dt1.predict(test_x)
```

```python
confusion_matrix(test_y,p2)
```

```python
accuracy_score(test_y,p2)*100
```
